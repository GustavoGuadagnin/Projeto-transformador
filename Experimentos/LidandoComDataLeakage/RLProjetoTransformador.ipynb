```python
delay_reason_cols = ['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']
df[delay_reason_cols] = df[delay_reason_cols].fillna(0)
df.head()
```

```python
df.drop(['CANCELLATION_REASON', 'FLIGHT_NUMBER', 'CANCELLED'], axis=1, inplace=True, errors='ignore')

df.drop('TAIL_NUMBER', axis=1, inplace=True)

df.drop(['ORIGIN_AIRPORT', 'DESTINATION_AIRPORT', 'AIRLINE'], axis=1, inplace=True, errors='ignore')
df = df.dropna(subset=['DEPARTURE_TIME','DEPARTURE_DELAY','TAXI_OUT','WHEELS_OFF','SCHEDULED_TIME','ELAPSED_TIME','AIR_TIME','WHEELS_ON','TAXI_IN','ARRIVAL_TIME','ARRIVAL_DELAY'])
```

```python
print(df.isna().sum() * 100 / len(df))
```

```python
df['DELAYED'] = df['ARRIVAL_DELAY'].apply(lambda x: 1 if x > 15 else 0)
```

```python
delay = df[df['DELAYED'] == 1]
on_time = df[df['DELAYED'] == 0]

delay_count = delay.shape[0]
on_time_count = on_time.shape[0]

total = delay_count + on_time_count
delay_percentage = (delay_count / total) * 100
on_time_percentage = (on_time_count / total) * 100

print(f"Não atrasados: {on_time_count:,}".replace(",", "."))
print(f"Atrasados: {delay_count:,}".replace(",", "."))
print(f"Percentual de voos não atrasados: {on_time_percentage:.2f}%")
print(f"Percentual de voos atrasados: {delay_percentage:.2f}%")
```

```python
from imblearn.under_sampling import RandomUnderSampler

X = df.drop(columns=['DELAYED'])
y = df['DELAYED']

rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X, y)


df_undersampled = pd.concat([X_resampled, y_resampled], axis=1)

on_time_count = (df_undersampled['DELAYED'] == 0).sum()
delay_count = (df_undersampled['DELAYED'] == 1).sum()
total = on_time_count + delay_count


on_time_percentage = (on_time_count / total) * 100
delay_percentage = (delay_count / total) * 100

print(f"Não atrasados: {on_time_count:,}".replace(",", "."))
print(f"Atrasados: {delay_count:,}".replace(",", "."))
print(f"Percentual de voos não atrasados: {on_time_percentage:.2f}%")
print(f"Percentual de voos atrasados: {delay_percentage:.2f}%")
```

# IMPLEMENTAÇÃO KNN

```python
df_delayed = df[df['DELAYED'] == 1]
df_on_time = df[df['DELAYED'] == 0]

sample_size_per_class = 1000000

df_delayed_sample = df_delayed.sample(n=sample_size_per_class, random_state=42)
df_on_time_sample = df_on_time.sample(n=sample_size_per_class, random_state=42)

df_sample = pd.concat([df_delayed_sample, df_on_time_sample])

df_sample = df_sample.sample(frac=1, random_state=42).reset_index(drop=True)

print("Shape do DataFrame original:", df.shape)
print("Shape do DataFrame amostrado:", df_sample.shape)
print("\Distribuição da classe 'DELAYED' no DataFrame amostrado:")
print(df_sample['DELAYED'].value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_sample = df_sample.drop(columns=['DELAYED'])
y_sample = df_sample['DELAYED']

X_train, X_test, y_train, y_test = train_test_split(
    X_sample, y_sample,
    test_size=0.3,
    stratify=y_sample,
    random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    max_depth=10,          
    criterion='gini',       
    min_samples_split=2,   
    min_samples_leaf=1,     
    random_state=42 
)
print("Configuração da Árvore de Decisão:")
print(f"• Profundidade máxima: {dt.max_depth}")
print(f"• Critério: {dt.criterion}")
print(f"• Min amostras divisão: {dt.min_samples_split}")
print(f"• Min amostras folha: {dt.min_samples_leaf}")

dt.fit(X_train, y_train)  # Árvore não precisa de normalização
```

## Avaliar o modelo Regressão Logística

### Subtask:
Fazer predições e avaliar o desempenho do modelo usando várias métricas.


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

y_train_pred = dt.predict(X_train_scaled)

y_pred = dt.predict(X_test_scaled)

train_accuracy = accuracy_score(y_train, y_train_pred)
train_precision = precision_score(y_train, y_train_pred)
train_recall = recall_score(y_train, y_train_pred)
train_f1 = f1_score(y_train, y_train_pred)

test_accuracy = accuracy_score(y_test, y_pred)
test_precision = precision_score(y_test, y_pred)
test_recall = recall_score(y_test, y_pred)
test_f1 = f1_score(y_test, y_pred)

cm_test = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm_test.ravel()

tnr = tn / (tn + fp)

print("=== MÉTRICAS DE AVALIAÇÃO - REGRESSÃO LOGÍSTICA ===")
print("\n### Métricas no TREINO ###")
print(f"Acurácia: {train_accuracy:.4f}")
print(f"Precisão: {train_precision:.4f}")
print(f"Recall (TPR): {train_recall:.4f}")
print(f"F1-score: {train_f1:.4f}")

print("\n### Métricas no TESTE ###")
print(f"Acurácia: {test_accuracy:.4f}")
print(f"Precisão: {test_precision:.4f}")
print(f"Recall (TPR): {test_recall:.4f}")
print(f"F1-score: {test_f1:.4f}")
print(f"True Negative Rate (TNR): {tnr:.4f}")

acc_diff = train_accuracy - test_accuracy
print(f"\n### Análise de Overfitting ###")
print(f"Diferença de acurácia (treino - teste): {acc_diff:.4f}")
if acc_diff > 0.05:
    print("⚠️  POSSÍVEL OVERFITTING DETECTADO!")
elif acc_diff < -0.05:
    print("⚠️  POSSÍVEL UNDERFITTING DETECTADO!")
else:
    print("✅ Modelo bem balanceado")

print(f"\n### Matriz de Confusão ###")
print(f"                 Predito")
print(f"                 0     1")
print(f"Real    0    {tn:5d} {fp:5d}")
print(f"        1    {fn:5d} {tp:5d}")

print(f"\n### Relatório de Classificação ###")
print(classification_report(y_test, y_pred, target_names=['Pontual', 'Atrasado']))
```

```python
import pandas as pd
import numpy as np

coefficients = dt.coef_[0]
intercept = dt.intercept_[0]
feature_names = X_train.columns

print("=== ANÁLISE DOS COEFICIENTES DA REGRESSÃO LOGÍSTICA ===")
print(f"\nIntercept (β₀): {intercept:.4f}")
print(f"\nCoeficientes das Features:")
print("-" * 70)

coef_df = pd.DataFrame({
    'Feature': feature_names,
    'Coeficiente': coefficients,
    'Abs_Coeficiente': np.abs(coefficients),
    'Odds_Ratio': np.exp(coefficients)
})

coef_df = coef_df.sort_values('Abs_Coeficiente', ascending=False)

print(f"{'Rank':<4} {'Feature':<25} {'Coeficiente':<12} {'Odds Ratio':<12} {'Impacto':<15}")
print("-" * 75)

for i, (_, row) in enumerate(coef_df.iterrows(), 1):
    coef = row['Coeficiente']
    odds_ratio = row['Odds_Ratio']
    
    # Determinando o impacto
    if coef > 0.1:
        impacto = "↑ Aumenta atraso"
    elif coef < -0.1:
        impacto = "↓ Reduz atraso"
    else:
        impacto = "≈ Neutro"
    
    print(f"{i:<4} {row['Feature']:<25} {coef:<12.4f} {odds_ratio:<12.4f} {impacto:<15}")

print(f"\n### TOP 5 FEATURES MAIS IMPORTANTES ###")
top_5 = coef_df.head(5)
for i, (_, row) in enumerate(top_5.iterrows(), 1):
    impact = "aumenta" if row['Coeficiente'] > 0 else "reduz"
    print(f"{i}. {row['Feature']}: {impact} probabilidade de atraso (coef: {row['Coeficiente']:.4f})")

print(f"\n### INTERPRETAÇÃO DOS COEFICIENTES ###")
print("• Coeficiente positivo: aumenta a probabilidade de atraso")
print("• Coeficiente negativo: reduz a probabilidade de atraso")
print("• Odds Ratio > 1: aumenta as chances de atraso")
print("• Odds Ratio < 1: reduz as chances de atraso")
print("• |Coeficiente| maior: maior importância da feature")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import cross_validate, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.metrics import make_scorer
import time

print("📊 COMPARAÇÃO: COM vs SEM DATA LEAKAGE")
print("=" * 50)


print("\n🚨 TREINANDO MODELO COM DATA LEAKAGE (Metodologia Incorreta)")
print("Usando TODAS as variáveis, incluindo informações futuras...")

X_with_leakage = df_sample.drop(columns=['DELAYED'])
y_with_leakage = df_sample['DELAYED']

X_train_leak, X_test_leak, y_train_leak, y_test_leak = train_test_split(
    X_with_leakage, y_with_leakage,
    test_size=0.3,
    stratify=y_with_leakage,
    random_state=42
)

scaler_leak = StandardScaler()
X_train_leak_scaled = scaler_leak.fit_transform(X_train_leak)
X_test_leak_scaled = scaler_leak.transform(X_test_leak)
lr_with_leakage = LogisticRegression(
    penalty='l2',
    C=1.0,
    solver='lbfgs',
    max_iter=1000,
    random_state=42
)

start_time = time.time()
lr_with_leakage.fit(X_train_leak_scaled, y_train_leak)
training_time_leak = time.time() - start_time

start_time = time.time()
y_pred_leak = lr_with_leakage.predict(X_test_leak_scaled)
prediction_time_leak = time.time() - start_time

acc_leak = accuracy_score(y_test_leak, y_pred_leak)
prec_leak = precision_score(y_test_leak, y_pred_leak)
rec_leak = recall_score(y_test_leak, y_pred_leak)
f1_leak = f1_score(y_test_leak, y_pred_leak)

print(f"✅ Modelo com data leakage treinado!")
print(f"Tempo de treinamento: {training_time_leak:.4f}s")
print(f"Tempo de predição: {prediction_time_leak:.4f}s")
```

```python

print("\n✅ TREINANDO MODELO SEM DATA LEAKAGE (Metodologia Correta)")
print("Usando apenas informações disponíveis ANTES do voo...")
safe_features = [
    'MONTH',                   
    'DAY',                    
    'DAY_OF_WEEK',
    'SCHEDULED_DEPARTURE',     
    'SCHEDULED_ARRIVAL',     
    'SCHEDULED_TIME',         
    'DISTANCE'               
]

available_safe_features = [f for f in safe_features if f in df_sample.columns]
print(f"Features seguras disponíveis: {len(available_safe_features)}")
print(f"Features: {available_safe_features}")


X_without_leakage = df_sample[available_safe_features]
y_without_leakage = df_sample['DELAYED']

X_train_safe, X_test_safe, y_train_safe, y_test_safe = train_test_split(
    X_without_leakage, y_without_leakage,
    test_size=0.3,
    stratify=y_without_leakage,
    random_state=42
)

scaler_safe = StandardScaler()
X_train_safe_scaled = scaler_safe.fit_transform(X_train_safe)
X_test_safe_scaled = scaler_safe.transform(X_test_safe)


lr_without_leakage = LogisticRegression(
    penalty='l2',
    C=1.0,
    solver='lbfgs',
    max_iter=1000,
    random_state=42
)

start_time = time.time()
lr_without_leakage.fit(X_train_safe_scaled, y_train_safe)
training_time_safe = time.time() - start_time

start_time = time.time()
y_pred_safe = lr_without_leakage.predict(X_test_safe_scaled)
prediction_time_safe = time.time() - start_time

acc_safe = accuracy_score(y_test_safe, y_pred_safe)
prec_safe = precision_score(y_test_safe, y_pred_safe)
rec_safe = recall_score(y_test_safe, y_pred_safe)
f1_safe = f1_score(y_test_safe, y_pred_safe)

print(f"✅ Modelo sem data leakage treinado!")
print(f"Tempo de treinamento: {training_time_safe:.4f}s")
print(f"Tempo de predição: {prediction_time_safe:.4f}s")
```

```python

print("\n" + "=" * 70)
print("COMPARAÇÃO: IMPACTO DA CORREÇÃO DE DATA LEAKAGE")
print("=" * 70)

print("\n### RESULTADOS COM DATA LEAKAGE (Metodologia Incorreta) ###")
print("⚠️ Usando TODAS as variáveis, incluindo informações futuras")
print(f"Features utilizadas: {X_with_leakage.shape[1]}")
print(f"Acurácia:  {acc_leak:.4f} ({acc_leak*100:.2f}%)")
print(f"Precisão:  {prec_leak:.4f}")
print(f"Recall:    {rec_leak:.4f}")
print(f"F1-Score:  {f1_leak:.4f}")

print("\n### RESULTADOS SEM DATA LEAKAGE (Metodologia Correta) ###")
print("✅ Usando apenas informações disponíveis ANTES do voo")
print(f"Features utilizadas: {X_without_leakage.shape[1]}")
print(f"Acurácia:  {acc_safe:.4f} ({acc_safe*100:.2f}%)")
print(f"Precisão:  {prec_safe:.4f}")
print(f"Recall:    {rec_safe:.4f}")
print(f"F1-Score:  {f1_safe:.4f}")

print("\n### IMPACTO DA CORREÇÃO ###")
acc_diff = acc_leak - acc_safe
rec_diff = rec_leak - rec_safe
f1_diff = f1_leak - f1_safe

print(f"📉 Redução na Acurácia:  {acc_diff:.4f} ({acc_diff*100:.2f} pontos percentuais)")
print(f"📉 Redução no Recall:    {rec_diff:.4f} ({rec_diff*100:.2f} pontos percentuais)")
print(f"📉 Redução no F1-Score:  {f1_diff:.4f} ({f1_diff*100:.2f} pontos percentuais)")

print("\n### INTERPRETAÇÃO ###")
if acc_leak > 0.95 and acc_safe < 0.70:
    print("🎯 CORREÇÃO BEM-SUCEDIDA!")
    print("✅ A redução dramática confirma que havia data leakage")
    print("✅ Os resultados corrigidos são mais realistas")
    print("✅ O modelo agora pode ser aplicado na prática")
elif acc_leak > 0.90:
    print("⚠️  POSSÍVEL DATA LEAKAGE DETECTADO")
    print("A acurácia muito alta sugere uso de informações futuras")
else:
    print("✅ RESULTADOS CONSISTENTES")
    print("Não há evidência forte de data leakage")

print("\n### FEATURES MAIS IMPORTANTES (SEM DATA LEAKAGE) ###")
if len(available_safe_features) > 0:
    feature_importance = abs(lr_without_leakage.coef_[0])
    feature_names = available_safe_features

    importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importancia': feature_importance
    }).sort_values('Importancia', ascending=False)
    
    for i, (_, row) in enumerate(importance_df.head(5).iterrows(), 1):
        print(f"{i}. {row['Feature']}: {row['Importancia']:.4f}")

print("\n### CONCLUSÃO ###")
print("📊 Esta comparação demonstra a importância de:")
print("1. ✅ Identificar e remover data leakage")
print("2. ✅ Usar apenas informações disponíveis antes da predição")
print("3. ✅ Aceitar resultados mais baixos mas realistas")
print("4. ✅ Validar a aplicabilidade prática do modelo")

if acc_safe >= 0.55:
    print(f"\n🎯 RESULTADO FINAL: {acc_safe:.1%} de acurácia é REALISTA para previsão de atrasos!")
else:
    print(f"\n⚠️  RESULTADO FINAL: {acc_safe:.1%} pode indicar necessidade de mais features ou dados")
```

```python


print("\n" + "=" * 60)
print("VALIDAÇÃO CRUZADA: COM vs SEM DATA LEAKAGE")
print("=" * 60)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score)
}


pipeline_with_leakage = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(
        penalty='l2',
        C=1.0,
        solver='lbfgs',
        max_iter=1000,
        random_state=42
    ))
])


pipeline_without_leakage = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(
        penalty='l2',
        C=1.0,
        solver='lbfgs',
        max_iter=1000,
        random_state=42
    ))
])

print("🔄 Executando validação cruzada COM data leakage...")
cv_results_leak = cross_validate(
    pipeline_with_leakage,
    X_with_leakage,
    y_with_leakage,
    cv=cv,
    scoring=scoring,
    return_train_score=True
)

print("🔄 Executando validação cruzada SEM data leakage...")
cv_results_safe = cross_validate(
    pipeline_without_leakage,
    X_without_leakage,
    y_without_leakage,
    cv=cv,
    scoring=scoring,
    return_train_score=True
)

print("✅ Validação cruzada concluída!")

print("\n### RESULTADOS DA VALIDAÇÃO CRUZADA ###")

print("\n📊 COM DATA LEAKAGE (Incorreto):")
print(f"Acurácia:  {cv_results_leak['test_accuracy'].mean():.4f} ± {cv_results_leak['test_accuracy'].std():.4f}")
print(f"Precisão:  {cv_results_leak['test_precision'].mean():.4f} ± {cv_results_leak['test_precision'].std():.4f}")
print(f"Recall:    {cv_results_leak['test_recall'].mean():.4f} ± {cv_results_leak['test_recall'].std():.4f}")
print(f"F1-Score:  {cv_results_leak['test_f1'].mean():.4f} ± {cv_results_leak['test_f1'].std():.4f}")

print("\n📊 SEM DATA LEAKAGE (Correto):")
print(f"Acurácia:  {cv_results_safe['test_accuracy'].mean():.4f} ± {cv_results_safe['test_accuracy'].std():.4f}")
print(f"Precisão:  {cv_results_safe['test_precision'].mean():.4f} ± {cv_results_safe['test_precision'].std():.4f}")
print(f"Recall:    {cv_results_safe['test_recall'].mean():.4f} ± {cv_results_safe['test_recall'].std():.4f}")
print(f"F1-Score:  {cv_results_safe['test_f1'].mean():.4f} ± {cv_results_safe['test_f1'].std():.4f}")

cv_leak_std = cv_results_leak['test_accuracy'].std()
cv_safe_std = cv_results_safe['test_accuracy'].std()

print("\n### ANÁLISE DE ESTABILIDADE ###")
print(f"Desvio padrão COM data leakage:  {cv_leak_std:.4f}")
print(f"Desvio padrão SEM data leakage:  {cv_safe_std:.4f}")

if cv_leak_std < 0.001:
    print("⚠️  Estabilidade SUSPEITA (muito baixa variação)")
else:
    print("✅ Estabilidade normal")

if cv_safe_std > 0.005:
    print("✅ Variação realista entre folds")
else:
    print("⚠️  Variação ainda baixa - investigar mais")
```
